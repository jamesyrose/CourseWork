```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx = pd.DataFrame()
wx = pd.DataFrame() 

def nx_func(x, p):
    return ((1-p) ** 2) * (p ** (x-1))

def wx_func(x, p): 
    return x * ((1-p) ** 2) * (p **(x - 1))


ps = [.9, .95, .99]

for p in ps: 
    nx[p] = [nx_func(a, p) for a in range(1, 301)]
    wx[p] = [wx_func(a, p) for a in range(1, 301)]

nx.plot(figsize=(10,10))
plt.title("Nx")
plt.xlabel("X")
plt.savefig("nx.png")

wx.plot(figsize=(10,10))
plt.title("Wx")
plt.xlabel("X")
plt.savefig("wx.png")
```

# Perdominate species 
Nx predominate species is 1 for all by virtue of the equation 
Nx = No(1-p)^2 * p^(x-1)

```python
# wx Predomeninate speices
print("p    Species")
print("-" * 20 )
for col in wx.columns: 
    print(f"{col}: {np.argmax(wx[col])  + 1}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt

def f(x): 
    return np.log(np.cosh(x))


def f_prime(x):
    return np.sinh(x) / np.cosh(x)


def q(x, y):
    return f(y) + f_prime(y)*(x-y) + .5 * (x-y)**2


rrange = np.linspace(-10, 10,100)

f_x = np.array(list(map(f, rrange)))
q1 = np.array(list(map(q, rrange, [-2] * len(rrange))))
q2 = np.array(list(map(q, rrange, [1] * len(rrange))))






plt.figure(figsize=(20,10))
plt.plot(rrange, f_x)
plt.plot(rrange, q1)
plt.plot(rrange, q2)
plt.legend(["f(x)", "Q(x, -2)", "Q(x, 1)"])



```

# Part f

```python
def update(x):
    # return  2 * np.sinh(x) / np.cosh(x) + 1 - np.tanh(x)*(-x ) -x
    return  x- np.tanh(x)
x = -2
res =[x]
for t in range(0, 10):
    x = update(x)
    res.append(x)
plt.figure()
plt.plot(res)
plt.title("x = -2")
    
x = 1
res =[x]
for t in range(0, 10):
    x = update(x)
    res.append(x)
plt.figure()
plt.plot(res)
plt.title("x = 1")
    
```

# Part g

```python
def update(x):
    # return  2 * np.sinh(x) / np.cosh(x) + 1 - np.tanh(x)*(-x ) -x
    return  x - np.tanh(x) / (1 - np.tanh(x)) ** 2
x = -2
res =[x]
for t in range(0, 25):
    x = update(x)
    res.append(x)
plt.figure()
plt.plot(res)
plt.title("x = -2")
    
x = 1
res =[x]
for t in range(0,100):
    x = update(x)
    res.append(x)
plt.figure()
plt.plot(res)
plt.title("x = 1")

    
```

# Part h

```python
def g(x):
    val = 0 
    for k in range(1, 11):
        val += np.log(np.cosh(x + 1/k))
    return val / 10

x = np.arange(-5, 5 , .1)
y = [g(val) for val in x]
plt.figure(figsize=(20,10))
plt.plot(x, y)
plt.title("g(x) = 1/10 * sum(log cosh(x + 1/k), k, 1, 10)")
    
```

Yes, it is still simple to extract the minima. The process to extract the minma is the same as the previous example. Using the fact we

know that f''(x) <= 1, we can inutively say that g''(x) is also <= 1. Thus the derivation for R(x,y) => g(x) is the same as Q(x,y) => f(x). 

In addition, the update function will be a slightly more complicated function (A little more time complexity), but the derivation remains

the same as Q(x,x_n). 


# Part k 

```python
def update(x):
    val = 0 
    for k in range(1, 11):
        val += np.tanh(x + 1/k)
    return x - 1/10 * val 

x = 0
l = [x]
while True:
    x = update(x)
    l.append(x)
    if (round(x, 4) == round(l[-2], 4)):
        break
plt.figure()
plt.plot(l)
plt.title(f"Minima at x_{len(l)} = {round(l[-1], 5)}")
```
